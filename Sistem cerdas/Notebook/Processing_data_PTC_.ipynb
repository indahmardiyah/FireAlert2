## analisis distribusi data

```python
#membuang fitur tahun hari bulan
df1 = df.drop(['day','month','year'], axis=1)
```

```python
# Encoding Not fire menjadi 0 and Fire menjadi 1
df1['Classes']= np.where(df1['Classes']== 'not fire',0,1)
```

```python
df1.head()
```

```python
df1.Classes.value_counts()
```

```python
# analisis Histogram utk memahami pola data
import seaborn as sns
sns.set_style("darkgrid")
df1.hist(bins=50, figsize=(20,15), ec = 'b')
plt.show()
```

```python
numeric_col = df1.select_dtypes(include=np.number).columns.tolist()
```

```python
# cari pola dan outlier
plt.style.use('ggplot')
plt.figure(figsize=(12, 8))
sns.boxplot(data=df[numeric_col])
plt.title('Boxplot fitur')
plt.show()
```

```python
percentage = df.Classes.value_counts(normalize=True)*100
percentage
```

**Charts**

```python
#plotting PieChart
classeslabels = ["FIRE", "NOT FIRE"]
plt.figure(figsize =(12, 7))
plt.pie(percentage,labels = classeslabels,autopct='%1.1f%%')
plt.title ("Pie Chart Classes", fontsize = 15)
plt.show()
```
